```python
overall_table=pd.DataFrame(columns=['chr','chromStart','chromEnd','name','strand','day1_fold_enri','day6_fold_enri','day18_fold_enri','day23_fold_enri','day27_fold_enri'])
```

```python
#overall_table
```

```python
#add all the records from all the time into the dataframe
time_course=[(day1, 'day1'),(day6, 'day6'), (day18, 'day18'), (day23, 'day23'), (day27,'day27')]
for time_point, time_point_name in time_course:
    for index, row in time_point.iterrows():
        overall_table.loc[len(overall_table),['chr','chromStart','chromEnd','name','strand']]=row[['chr','chromStart','chromEnd','name','strand']]
        overall_table.ix[len(overall_table)-1,['day1_fold_enri','day6_fold_enri','day18_fold_enri','day23_fold_enri','day27_fold_enri']]=np.zeros(5)+1
        overall_table.ix[len(overall_table)-1, time_point_name+'_fold_enri']=row.fold_enrchment
    #print row[['chr','chromStart','chromEnd','name','strand','fold_enrchment']]
```

```python
#overall_table
```

```python
#Check redundance of the table again, uniform the coordinates of the records with 85% sharing loci region.
#stored the new dataframe in overall_table2
##steps:1.group the records by 'name', for each group look for those records with similar coordinates (sharing >85% loci region),and convert their coordinates to the same, stored in new dataframe.
##      2.group the records of the new dataframe by 'chr','chromStart','chromEnd','name', for each group get a representative record by average all the records, ps: those record item value with 1 should not be involved in the average. 
grouped=overall_table.groupby('name')
overall_table2=pd.DataFrame(columns=overall_table.columns)
for name, group in grouped:
    for ind, row in group.iterrows():
        for ind2, row2 in group.iterrows():
            if(ind==ind2):
                continue
            if(len(set(range(row.chromStart,row.chromEnd)).intersection(set(range(row2.chromStart,row2.chromEnd))))*1.0/min(len(set(range(row.chromStart,row.chromEnd))),len(set(range(row2.chromStart,row2.chromEnd))))>0.85):
                row.chromStart=row2.chromStart=min(row.chromStart, row2.chromStart)
                row.chromEnd=row2.chromEnd=max(row.chromEnd, row2.chromEnd)
    #print name
    for ind, row in group.iterrows():
        overall_table2.loc[len(overall_table2)]=row

#combine the records with same/similar loci coordinates, and store the new dataframe in overall_table3
grouped2=overall_table2.groupby(['chr','chromStart','chromEnd', 'name'])
overall_table3=pd.DataFrame(columns=overall_table.columns)
for name,group in grouped2:
    if (len(group)==1):
        overall_table3.loc[len(overall_table3)]=group.iloc[0]
        continue
    day1_sum=0
    day1_count=0
    day6_sum=0
    day6_count=0
    day18_sum=0
    day18_count=0
    day23_sum=0
    day23_count=0
    day27_sum=0
    day27_count=0
    for ind, row in group.iterrows():
        if(row.day1_fold_enri!=1):
            day1_sum=day1_sum+row.day1_fold_enri
            day1_count=day1_count+1
        if(row.day6_fold_enri!=1):
            day6_sum=day6_sum+row.day6_fold_enri
            day6_count=day6_count+1
        if(row.day18_fold_enri!=1):
            day18_sum=day18_sum+row.day18_fold_enri
            day18_count=day18_count+1
        if(row.day23_fold_enri!=1):
            day23_sum=day23_sum+row.day23_fold_enri
            day23_count=day23_count+1
        if(row.day27_fold_enri!=1):
            day27_sum=day27_sum+row.day27_fold_enri
            day27_count=day27_count+1
    tmp=group.iloc[0]
    if(day1_count==0):
        tmp.day1_fold_enri=1
    else:
        tmp.day1_fold_enri=day1_sum*1.0/day1_count
    if(day6_count==0):
        tmp.day6_fold_enri=1
    else:
        tmp.day6_fold_enri=day6_sum*1.0/day6_count
    if(day18_count==0):
        tmp.day18_fold_enri=1
    else:
        tmp.day18_fold_enri=day18_sum*1.0/day18_count
    if(day23_count==0):
        tmp.day23_fold_enri=1
    else:
        tmp.day23_fold_enri=day23_sum*1.0/day23_count
    if(day27_count==0):
        tmp.day27_fold_enri=1
    else:
        tmp.day27_fold_enri=day27_sum*1.0/day27_count
    
    overall_table3.loc[len(overall_table3)]=tmp
    
overall_table3
```
